```python
def remark(x):
    if x < 30: return "good"
    if 30 < x < 60: return "satisfactory"
    if 60 < x < 90: return "moderate"
    if 90 < x < 120: return "poor"
    if 120 < x < 250: return "very poor"
    else: return "severe"
    
data['remark'] = [remark(x) for x in data['PM2.5']]
```

```python
data.head()
```

```python
plt.figure(figsize=(9, 5))
sns.stripplot(x="Year", y="PM2.5", data=data, hue="remark")
```
